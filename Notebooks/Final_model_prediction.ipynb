```python
# Merge Function
def merge(Cropset, name):
    Cropset["Name"] = Cropset["Name"].str.title()
    Geospatial_dB["COUNTY"] = Geospatial_dB["COUNTY"].str.title()
    Cropset = Cropset.rename(columns={'Name':'COUNTY'}, inplace=False)
    final = Cropset.merge(Geospatial_dB,on="COUNTY", #on=['Year', 'COUNTY']
                         )
    # Save the dataset
    final.to_csv((os.path.join(home, "final_dataset", f"{name}.csv")))
    return final
```

```python
corn = merge(Corn, "Corn")
soy = merge(Soy, "Soybean")
cotton = merge(Cotton, "Cotton")
wheat = merge(Wheat, "Wheat")
peanut = merge(Peanut, "Peanut")
len(corn.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize features (important for PCA)
# Drop commodity and irrelevant columns
corn["Yield"] = corn["Yield"].fillna(0) + corn["Silage_Yield_LBS_PER_ACRE"].fillna(0)
corn_feature = corn.drop(columns="Commodity", errors="ignore")
corn_feature.drop(
    ['Program', 'Period', 'Week Ending', 'Geo Level', 'State', 'Domain Category'],
    axis=1, errors="ignore", inplace=True
)
corn_feature.drop(columns=def_delete, errors="ignore", inplace=True)

# 0️⃣ Drop columns with any NaN
corn_feature = corn_feature.dropna(axis=1)
#print("Columns remaining after dropping NaN columns (corn):", corn_feature.columns.tolist())

# 1️⃣ Drop highly correlated features
numeric_features_corn = corn_feature.select_dtypes(include=["number"])
corr_matrix_corn = numeric_features_corn.corr().abs()
upper_corn = corr_matrix_corn.where(np.triu(np.ones(corr_matrix_corn.shape), k=1).astype(bool))
to_drop_corn = [col for col in upper_corn.columns if any(upper_corn[col] > 0.9)]
print("Highly correlated features to drop (corn):", to_drop_corn, "Length:", len(to_drop_corn))

corn_feature_reduced = corn_feature.drop(columns=to_drop_corn)

# 2️⃣ Encode county
features_corn = corn_feature_reduced.copy()
le_corn = LabelEncoder()
features_corn["county_encoded_corn"] = le_corn.fit_transform(features_corn["COUNTY"])
features_corn = features_corn.drop(columns="COUNTY")

# 3️⃣ Separate predictors and target
X_corn = features_corn.drop(columns="Yield")
y_corn = features_corn["Yield"]

# 4️⃣ Scale numeric features except county_encoded
#numeric_cols_corn = X_corn.select_dtypes(include=["number"]).columns.tolist()
#cols_to_scale_corn = [col for col in numeric_cols_corn if col != "county_encoded_corn"]
#scaler_corn = StandardScaler()
#X_corn[cols_to_scale_corn] = scaler_corn.fit_transform(X_corn[cols_to_scale_corn])

# 5️⃣ Create pseudo-stratification label (year + county)
X_corn['year_county_strat_corn'] = features_corn['Year'].astype(str)
X_corn = X_corn.drop(columns=['Year'], errors='ignore')
# + "_" + features_corn['county_encoded_corn'].astype(str)  # optional

# 6️⃣ Stratified K-Fold
skf_corn = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
r2_scores_corn = []
importances_folds_corn = []

for train_idx, test_idx in skf_corn.split(X_corn, X_corn['year_county_strat_corn']):
    X_train_corn = X_corn.iloc[train_idx].drop(columns=['year_county_strat_corn'])
    X_test_corn = X_corn.iloc[test_idx].drop(columns=['year_county_strat_corn'])
    y_train_corn, y_test_corn = y_corn.iloc[train_idx], y_corn.iloc[test_idx]

    rf_corn = RandomForestRegressor(n_estimators=500, random_state=42)
    rf_corn.fit(X_train_corn, y_train_corn)

    r2_scores_corn.append(rf_corn.score(X_test_corn, y_test_corn))
    importances_folds_corn.append(rf_corn.feature_importances_)

# 7️⃣ Average feature importance across folds
avg_importance_corn = np.mean(importances_folds_corn, axis=0)
importance_avg_corn = pd.Series(avg_importance_corn, index=X_train_corn.columns).sort_values(ascending=False).head(20)

# 8️⃣ Print R²
print("R² scores per fold (corn):", r2_scores_corn)
print("Mean R² (corn):", np.mean(r2_scores_corn))

# 9️⃣ Plot top 20 feature importance
fig, ax = plt.subplots(figsize=(10, 6))
importance_avg_corn.plot(kind="barh", ax=ax)
ax.invert_yaxis()
ax.set_xlabel("Average Feature Importance (RF)")
ax.set_title("Top 20 Features Influencing Corn Yield")
plt.tight_layout()
fig.savefig("features_corn.png", dpi=300, bbox_inches="tight")
plt.show()

# 10️⃣ Train final model on all corn data
final_rf_corn = RandomForestRegressor(n_estimators=500, random_state=42)
final_rf_corn.fit(X_corn.drop(columns=['year_county_strat_corn']), y_corn)
```

```python
def predict_crop_yield(model, df_2023, le, X_train_cols, to_drop, scaler=None, county_col='COUNTY', encoded_col='county_encoded'):
    # Drop irrelevant columns
    df = df_2023.drop(columns="Commodity", errors="ignore")
    df.drop(['Program', 'Period', 'Week Ending', 'Geo Level', 'State', 'Domain Category'],
            axis=1, errors="ignore", inplace=True)
    df.drop(columns=def_delete, errors="ignore", inplace=True)
    df = df.drop(columns=to_drop, errors="ignore")
    
    # Keep only counties seen during training
    df = df[df[county_col].isin(le.classes_)].copy()
    
    # Encode county
    df[encoded_col] = le.transform(df[county_col])
    df.drop(columns=county_col, inplace=True, errors="ignore")
    
    # Align columns to training set
    X_2023 = df.reindex(columns=X_train_cols, fill_value=0)
    
    # Scale numeric features if scaler is provided
    if scaler is not None:
        numeric_cols = X_2023.select_dtypes(include=["number"]).columns.tolist()
        cols_to_scale = [col for col in numeric_cols if col != encoded_col]
        #X_2023[cols_to_scale] = scaler.transform(X_2023[cols_to_scale])
    
    # Predict
    df["Predicted_Yield"] = model.predict(X_2023)
    return df
```
